## Import libraries and data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from scipy.stats import poisson
import math
```

```python
data = pd.read_csv('../data/NBA_all_results.csv')
```

## Count match totals

```python
data['total_points'] = data['PTS'] + data['PTS.1']
histo = data.groupby(by = 'total_points')['Date'].count()
histo = pd.DataFrame(histo)
histo.columns = ['count']
histo = histo.reindex(list(range(134,291,1)))
histo['total_points'] = histo.index
histo = histo.fillna(0)
histo['prob'] = histo['count'] / histo['count'].sum()
plt.scatter(histo.total_points , histo['count'])
```

## Modelling

```python
def negbin(mean,R,x):
    p = (mean / R) * (1/(1+(mean/R)))
    return ((math.factorial(x + R -1))/(math.factorial(x)*math.factorial(R -1))) * (p ** x) * ((1-p)**R)

```

```python
mean = 207.18 #Define lambda 
R = 280 # Define R value

histo['model'] = histo.apply(lambda x: negbin(mean , R , x['total_points']) , axis = 1)

plt.plot(histo.total_points, histo.model, 'b')
plt.plot(histo.total_points , histo['prob'],'r.')
plt.show
```

## Chi Squared fitting

```python
def testing(R):
    histo['model'] = histo.apply(lambda x: negbin(207.18 , R , x['total_points']) , axis = 1)
    histo['chi'] = ((histo.prob - histo.model)**2)/histo.model
    return histo.chi.sum()

R_value = pd.DataFrame(data =({'R' : list(range(100,350,1)) , 'Sum_Chi' : list(range(250))}))

R_value.Sum_Chi = R_value.apply(lambda x: testing(x['R']), axis=1)

R_value.R[R_value.Sum_Chi.idxmin()] # Returns optimal value of R that minimises sum error
    
    
```

## Home and Away Rankings

```python
away = data.groupby(by = 'Visitor/Neutral')['PTS'].sum()
away = pd.DataFrame(away)
away['games'] = data.groupby(by='Visitor/Neutral')['Date'].count()
away['Offence'] = away.PTS / away.games
away['PTS.1'] = data.groupby(by='Visitor/Neutral')['PTS.1'].sum()
away['Defence'] = away['PTS.1'] / away.games
away
```

```python
home = data.groupby(by = 'Home/Neutral')['PTS.1'].sum()
home = pd.DataFrame(home)
home['games'] = data.groupby(by='Home/Neutral')['Date'].count()
home['Offence'] = home['PTS.1'] / home.games
home['PTS'] = data.groupby(by='Home/Neutral')['PTS'].sum()
home['Defence'] = home['PTS'] / home.games
home
```
